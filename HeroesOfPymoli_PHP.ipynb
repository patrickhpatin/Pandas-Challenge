```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)

purchase_data
# print(purchase_data.head())
```

## Player Count

* Display the total number of players


```python
#  Unique Player Count
Player_Count = len(purchase_data["SN"].unique())

print(f"Number of Players = {Player_Count}")
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
# Create a groupby dataframe with unique products
Unique_Items = purchase_data.groupby(["Item ID", "Price"])

# Get the number of unique items in our dataset
Item_Count = len(Unique_Items)

# Calculate the average price of our unique items and format to currency
Avg_Price = '${:0,.2f}'.format(purchase_data["Price"].mean())

# Get the total number of purchases (rows) in our dataframe
Purchase_Count = len(purchase_data)

# Total revenue is the sum of all purchase data (not just unique records)
Total_Revenue = '${:0,.2f}'.format(purchase_data["Price"].sum())

# Create our dictionary
my_data = {'# Unique Items': [Item_Count], 
           'Average Price': [Avg_Price],
           '# Purchases': [Purchase_Count], 
           'Total Revenue': [Total_Revenue]}
  
# Create the pandas DataFrame out of our dictionary
my_df = pd.DataFrame(my_data, columns = ['# Unique Items', 'Average Price',  '# Purchases', 'Total Revenue'])

my_df
# print(my_df.head())
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
# Create a dataframe of unique Screen Name records
Unique_Players = purchase_data.drop_duplicates("SN")

dfGenders = Unique_Players.groupby(["Gender"])["SN"].count().to_frame()
dfGenders = dfGenders.rename(columns = {"SN":'Total Count'})

dfGenders["Percent"] = dfGenders["Total Count"] / len(Unique_Players) * 100
dfGenders["Percent"] = dfGenders["Percent"].astype(float).map('{:,.2f}%'.format)

dfGenders
# print(dfGenders.head())
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# Purchase Counts by Gender
Purchases_Combined = purchase_data.groupby(["Gender"])["SN"].count().to_frame()
Purchases_Combined = Purchases_Combined.rename(columns = {"SN":'Purchase Count'})

# Average Purchase Price by Gender
Purchases_Combined["Average Purchase"] = purchase_data.groupby(["Gender"])["Price"].mean()

# Total Purchase Value by Gender
Purchases_Combined["Total Purchase Value"] = purchase_data.groupby(["Gender"])["Price"].sum()

# Average Purchases by Screen Name
Purchases_Combined["Avg Total Purchase per Person"] = Purchases_Combined["Total Purchase Value"] / dfGenders["Total Count"]

# Now Format My Currency Columns
Purchases_Combined["Avg Total Purchase per Person"] = Purchases_Combined["Avg Total Purchase per Person"].astype(float).map('${:,.2f}'.format)
Purchases_Combined["Total Purchase Value"] = Purchases_Combined["Total Purchase Value"].astype(float).map('${:,.2f}'.format)
Purchases_Combined["Average Purchase"] = Purchases_Combined["Average Purchase"].astype(float).map('${:,.2f}'.format)

Purchases_Combined
# print(Purchases_Combined.head())
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# Create the bins in which Data will be held
bins = [0, 9, 14, 19, 24, 29, 34, 39, 1000]

# Create the names for the four bins
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

# Create a dataframe of unique Screen Name records
Unique_Players = purchase_data.drop_duplicates(["SN"])
Total_Records = len(Unique_Players)

# Calculate the percentage of each age group
purchase_data["Bins"] = pd.cut(Unique_Players["Age"], bins, labels=group_names)
Age_Bins = purchase_data.groupby(["Bins"])["SN"].count().to_frame()
Age_Bins["Percentage of Players"] = Age_Bins["SN"] / Total_Records * 100

# Column formatting & renaming
Age_Bins = Age_Bins.rename(columns = {"SN":'Total Count'})
Age_Bins["Percentage of Players"] = Age_Bins["Percentage of Players"].astype(float).map('{:,.2f}%'.format)

Age_Bins
# print(Age_Bins.head())
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# We will be using the same variables (bins and group_names) as the last section
# Reload the data file
purchase_data = pd.read_csv(file_to_load)
purchase_data["Bins"] = pd.cut(purchase_data["Age"], bins, labels=group_names)

# Create a dataframe of unique Screen Name records
AgeGroup_PurchSums = purchase_data.groupby(["Bins"])["Price"].sum().to_frame()
AgeGroup_PurchSums["Price"] = AgeGroup_PurchSums["Price"].astype(float).map('{:,.2f}'.format)

AgeGroup_PurchCounts = purchase_data.groupby(["Bins"])["Item ID"].count().to_frame()
AgeGroup_PurchCounts["Item ID"] = AgeGroup_PurchCounts["Item ID"].astype(float).map('{:,.2f}'.format)

# Purchase Count
Purchase_Bins = purchase_data.groupby(["Bins"])["Item ID"].count().to_frame()
Purchase_Bins.reset_index(level=0, inplace=True)
Purchase_Bins = Purchase_Bins.rename(columns = {"Bins":'Age Ranges', "Item ID": "Purchase Count"})

# Average Purchase Price
Purchase_Sums = purchase_data.groupby(["Bins"])["Price"].sum().to_frame()
Purchase_Sums.reset_index(level=0, inplace=True)
Purchase_Counts = purchase_data.groupby(["Bins"])["Price"].count().to_frame()
Purchase_Counts.reset_index(level=0, inplace=True)
Purchase_Bins["Average Purchase Price"] = Purchase_Sums["Price"] / Purchase_Counts["Price"]

# Total Purchase Value
Purchase_Bins["Total Purchase Value"] = Purchase_Sums["Price"]

# Avg Total Purchase per Person
Age_Bins = Age_Bins.rename(columns = {"Bins":'Age Ranges'})
Age_Bins.reset_index(level=0, inplace=True)
Purchase_Bins["Avg Total Purchase per Person"] = Purchase_Bins["Total Purchase Value"] / Age_Bins["Total Count"]

# Formatting
Purchase_Bins["Total Purchase Value"] = Purchase_Bins["Total Purchase Value"].astype(float).map('${:,.2f}'.format)
Purchase_Bins["Average Purchase Price"] = Purchase_Bins["Average Purchase Price"].astype(float).map('${:,.2f}'.format)
Purchase_Bins["Avg Total Purchase per Person"] = Purchase_Bins["Avg Total Purchase per Person"].astype(float).map('${:,.2f}'.format)

Purchase_Bins
# print(Purchase_Bins.head())
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Aggregate our data
Top_Spenders = purchase_data.groupby(["SN"]).agg({'Price': ['count', 'mean', 'sum']})
Top_Spenders.columns = ['_'.join(col) for col in Top_Spenders.columns.values]
Top_Spenders = Top_Spenders.rename(columns = {"Price_count":"Purchase Count", "Price_mean": "Average Purchase Price", "Price_sum": "Total Purchase Value"})
Top_Spenders = Top_Spenders.sort_values('Total Purchase Value', ascending=False)

# Formatting
Top_Spenders["Total Purchase Value"] = Top_Spenders["Total Purchase Value"].astype(float).map('${:,.2f}'.format)
Top_Spenders["Average Purchase Price"] = Top_Spenders["Average Purchase Price"].astype(float).map('${:,.2f}'.format)

Top_Spenders.head()
# print(Top_Spenders.head())
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Aggregate our data
Pop_Items = purchase_data.groupby(["Item ID", "Item Name"]).agg({'Price': ['count', 'min', 'sum']})
Pop_Items.columns = ['_'.join(col) for col in Pop_Items.columns.values]
Pop_Items = Pop_Items.rename(columns = {"Price_count":"Purchase Count", "Price_min": "Item Price", "Price_sum": "Total Purchase Value"})
Pop_Items = Pop_Items.sort_values('Purchase Count', ascending=False)

# Before converting the numeric columns to strings (i.e. formatting), copy this for the next step to save time
Prof_Items = Pop_Items.copy()

# Formatting
Pop_Items["Item Price"] = Pop_Items["Item Price"].astype(float).map('${:,.2f}'.format)
Pop_Items["Total Purchase Value"] = Pop_Items["Total Purchase Value"].astype(float).map('${:,.2f}'.format)

Pop_Items.head()
# print(Pop_Items.head())
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
# Sort our dataframe, it has already been aggregated
Prof_Items = Prof_Items.sort_values('Total Purchase Value', ascending=False)

# Formatting
Prof_Items["Item Price"] = Prof_Items["Item Price"].astype(float).map('${:,.2f}'.format)
Prof_Items["Total Purchase Value"] = Prof_Items["Total Purchase Value"].astype(float).map('${:,.2f}'.format)

Prof_Items.head()
# print(Prof_Items.head())
```
